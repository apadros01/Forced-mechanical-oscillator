```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_points(file_path):
    xs=[]
    ys=[]
    with open(file_path) as f:
        for line in f:
            x, y = line.split()
            xs.append(x)
            ys.append(y)
    xs=np.array(xs).astype(float)
    ys=np.array(ys).astype(float)
    return xs,ys
    
```

# Varying the value of w

```python

x_ress,y_ress= get_points('ressonance/RK4-points_x.txt')
x_11,y_11 = get_points('w_11/RK4-points_x.txt')
x_12,y_12 = get_points('w_12/RK4-points_x.txt')
```

```python
plt.plot(x_ress[0:500],y_ress[0:500],color='black',label='$\omega = \omega_0$')
plt.plot(x_11[0:500],y_11[0:500],color='orange',label='$\omega = 1.1\omega_0$')
plt.plot(x_12[0:500],y_12[0:500],color='blue',label='$\omega = 1.2\omega_0$')
plt.legend()
plt.show()
```
